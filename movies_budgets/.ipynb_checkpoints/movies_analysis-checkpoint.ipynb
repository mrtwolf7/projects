```python
import pandas as pd

df = pd.read_csv("movie-budgets-enriched.csv")

#df.dtypes

df['profit_ratio'] = df['worldwide_gross']/df['production_budget']
bin_edges = list(range(1920, 2031, 10))
theatres_bins = [0, 10, 500, 3000, np.inf]
duration_bins = [0, 75, 90, 120, 150, np.inf]
bin_labels = ['1920s','1930s','1940s','1950s','1960s','1970s','1980s', '1990s', '2000s', '2010s', '2020s']
theater_labels = ['Ultra-Limited', 'Limited', 'Wide', 'Ultra-Wide']
duration_labels = ['Short', 'Concise', 'Standard', 'Extended', 'Epic']

df['decade'] = pd.cut(df['year'], bins=bin_edges, labels=bin_labels, right=False)
df['theatres_open_bin'] = pd.cut(df['opening_theater_count'], bins=theatres_bins, labels=theater_labels, right=False)
df['durations'] = pd.cut(df['runtime'], bins=duration_bins, labels=duration_labels, right=False)

df.head(5)
```

### Predicting bins

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.utils.class_weight import compute_sample_weight

# --- 1. DATA CLEANING ---
genre_counts = df['genre'].value_counts()
minor_genres = genre_counts[genre_counts < 50].index
df['genre_clean'] = df['genre'].replace(minor_genres, 'Other')
df_modern = df[df['year'] >= 1990].copy()

# --- 2. THE NEW 3-TIER TARGET DEFINITION ---
# 0 = Loss/Underperformed (<2x)
# 1 = Profitable (2x - 5x)
# 2 = Mega-Hit (>5x)
bin_edges = [0, 2.0, 5.0, np.inf]
bin_labels = [0, 1, 2]  

df_modern['target_bin'] = pd.cut(df_modern['profit_ratio'], bins=bin_edges, labels=bin_labels, right=False)
df_modern['target_bin'] = df_modern['target_bin'].astype(int)

# --- 3. BASIC FEATURE ENGINEERING ---
df_modern['log_budget'] = np.log1p(df_modern['production_budget'])
df_modern['month_sin'] = np.sin(2 * np.pi * df_modern['month'] / 12)
df_modern['month_cos'] = np.cos(2 * np.pi * df_modern['month'] / 12)

train_mask = df_modern['year'] < 2015
test_mask = df_modern['year'] >= 2015

# --- 4. HANDLE MISSING VALUES (IMPUTATION) ---
theater_median = df_modern.loc[train_mask, 'opening_theater_count'].median()
runtime_median = df_modern.loc[train_mask, 'runtime'].median()

df_modern['opening_theater_count'] = df_modern['opening_theater_count'].fillna(theater_median)
df_modern['runtime'] = df_modern['runtime'].fillna(runtime_median)

# --- 5. ORDINAL ENCODING FOR DURATIONS (1 Feature!) ---
duration_bins = [0, 75, 90, 120, 150, np.inf]
# Assign escalating numbers instead of text so it stays in ONE column
duration_labels_ordinal = [0, 1, 2, 3, 4] 
df_modern['duration_ordinal'] = pd.cut(df_modern['runtime'], bins=duration_bins, labels=duration_labels_ordinal, right=False)
df_modern['duration_ordinal'] = df_modern['duration_ordinal'].astype(int)

# --- 6. EFFICIENCY FEATURE ENGINEERING ---
df_modern['log_opening_theaters'] = np.log1p(df_modern['opening_theater_count'])
df_modern['log_budget_per_theater'] = np.log1p(df_modern['production_budget'] / (df_modern['opening_theater_count'] + 1))

# --- 7. TARGET ENCODING (GENRE) ---
# Profitable is now defined as bin 1 or higher in the new 3-tier system
df_modern['is_profitable'] = (df_modern['target_bin'] >= 1).astype(int)
genre_train_prof_rates = df_modern[train_mask].groupby('genre_clean')['is_profitable'].mean()
df_modern['genre_encoded'] = df_modern['genre_clean'].map(genre_train_prof_rates)
df_modern['genre_encoded'] = df_modern['genre_encoded'].fillna(df_modern[train_mask]['is_profitable'].mean())

# --- 8. BUDGET TO GENRE INTERACTION ---
genre_median_budgets = df_modern[train_mask].groupby('genre_clean')['production_budget'].median()
df_modern['genre_median_budget'] = df_modern['genre_clean'].map(genre_median_budgets).fillna(df_modern[train_mask]['production_budget'].median())
df_modern['log_budget_to_genre_ratio'] = np.log1p(df_modern['production_budget'] / df_modern['genre_median_budget'])

# --- 9. DEFINE FEATURES LIST ---
features = [
    'log_budget', 
    'month_sin', 
    'month_cos', 
    'genre_encoded', 
    'log_budget_to_genre_ratio',
    'log_opening_theaters',
    'log_budget_per_theater',
    'duration_ordinal' 
]

X_train = df_modern.loc[train_mask, features]
y_train = df_modern.loc[train_mask, 'target_bin']

X_test = df_modern.loc[test_mask, features]
y_test = df_modern.loc[test_mask, 'target_bin']

# --- 10. TRAIN GRADIENT BOOSTING ---
train_sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

clf = GradientBoostingClassifier(
    n_estimators=200, 
    learning_rate=0.1, 
    max_depth=5, 
    random_state=42
)
clf.fit(X_train, y_train, sample_weight=train_sample_weights)

# --- 11. EVALUATE PERFORMANCE ---
y_pred = clf.predict(X_test)

print(f"--- 3-Tier Gradient Boosting Evaluation ---")
print(f"Overall Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
print("Classification Report:")
target_names = ['Loss / Underperformed (<2x)', 'Profitable (2x-5x)', 'Mega-Hit (>5x)']
print(classification_report(y_test, y_pred, target_names=target_names))

print("\n--- Feature Importances ---")
importances = pd.Series(clf.feature_importances_, index=features).sort_values(ascending=False)
print(importances)
```

## Box office prediction

When building the box office prediction, we have to take into account some biases we have observed in the data. Probably the number of movies we have for minor genres are not enough to build a reliable predictor, same as the decade, where mot movies of the dataset are from 00s onwards.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- 1. DATA CLEANING ---
genre_counts = df['genre'].value_counts()
minor_genres = genre_counts[genre_counts < 50].index
df['genre_clean'] = df['genre'].replace(minor_genres, 'Other')

df_modern = df[df['year'] >= 1990].copy()
df_modern['profit_ratio'] = df_modern['profit_ratio'].clip(upper=20)

# --- 2. BASIC FEATURE ENGINEERING (Keep this the same) ---
df_modern['log_budget'] = np.log1p(df_modern['production_budget'])
df_modern['month_sin'] = np.sin(2 * np.pi * df_modern['month'] / 12)
df_modern['month_cos'] = np.cos(2 * np.pi * df_modern['month'] / 12)

# Create the target column
df_modern['target'] = np.log1p(df_modern['profit_ratio'])

# --- 3. TIME-BASED MASKS (Define train vs test rows first) ---
# We must create the masks here so we know which rows belong to the training set
train_mask = df_modern['year'] < 2015
test_mask = df_modern['year'] >= 2015


# =====================================================================
# TARGET ENCODING BLOCK
# =====================================================================

# Step A: Calculate the average target value per genre USING ONLY THE TRAIN DATA
genre_train_means = df_modern[train_mask].groupby('genre_clean')['target'].mean()

# Step B: Map those averages back to the main dataframe
df_modern['genre_encoded'] = df_modern['genre_clean'].map(genre_train_means)

# Step C: Handle any potential missing values 
global_train_mean = df_modern[train_mask]['target'].mean()
df_modern['genre_encoded'] = df_modern['genre_encoded'].fillna(global_train_mean)

# Step D: Define your new feature list
features = ['log_budget', 'month_sin', 'month_cos', 'genre_encoded']

# =====================================================================


# --- 4. SPLIT DATA INTO X AND Y (Updated) ---
X_train = df_modern.loc[train_mask, features]
y_train = df_modern.loc[train_mask, 'target']

X_test = df_modern.loc[test_mask, features]
y_test = df_modern.loc[test_mask, 'target']

print(f"Training samples: {X_train.shape[0]} | Testing samples: {X_test.shape[0]}")

# --- 5. MODEL TRAINING & EVALUATION (Keep this the same ---
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred_log = model.predict(X_test)

y_test_original = np.expm1(y_test)
y_pred_original = np.expm1(y_pred_log)

print(f"\n--- Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test_original, y_pred_original):.2f}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test_original, y_pred_original)):.2f}")
print(f"R² Score (Log-scale): {r2_score(y_test, y_pred_log):.4f}")

print("\n--- Feature Importances ---")
importances = pd.Series(model.feature_importances_, index=features).sort_values(ascending=False)
print(importances)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor  # <-- Switched to Gradient Boosting
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- 1. DATA CLEANING ---
genre_counts = df['genre'].value_counts()
minor_genres = genre_counts[genre_counts < 50].index
df['genre_clean'] = df['genre'].replace(minor_genres, 'Other')

df_modern = df[df['year'] >= 1990].copy()
df_modern['profit_ratio'] = df_modern['profit_ratio'].clip(upper=20)

# --- 2. FEATURE ENGINEERING & BUG FIX ---
df_modern['log_budget'] = np.log1p(df_modern['production_budget'])

# FIX: One is sin, one is cos
df_modern['month_sin'] = np.sin(2 * np.pi * df_modern['month'] / 12)
df_modern['month_cos'] = np.cos(2 * np.pi * df_modern['month'] / 12)

# Create the target column
df_modern['target'] = np.log1p(df_modern['profit_ratio'])

# Define train vs test rows first
train_mask = df_modern['year'] < 2015
test_mask = df_modern['year'] >= 2015

# --- 3. TARGET ENCODING (GENRE) ---
genre_train_means = df_modern[train_mask].groupby('genre_clean')['target'].mean()
df_modern['genre_encoded'] = df_modern['genre_clean'].map(genre_train_means)
global_train_mean = df_modern[train_mask]['target'].mean()
df_modern['genre_encoded'] = df_modern['genre_encoded'].fillna(global_train_mean)

# --- 4. ADVANCED FEATURE: BUDGET RELATIVE TO GENRE BASELINE ---
# Find the median budget for each genre in the training set
genre_median_budgets = df_modern[train_mask].groupby('genre_clean')['production_budget'].median()
df_modern['genre_median_budget'] = df_modern['genre_clean'].map(genre_median_budgets).fillna(df_modern[train_mask]['production_budget'].median())

# Calculate how many times bigger/smaller this movie's budget is compared to its genre norm
df_modern['budget_to_genre_ratio'] = df_modern['production_budget'] / df_modern['genre_median_budget']
# Log-transform it to keep scales manageable
df_modern['log_budget_to_genre_ratio'] = np.log1p(df_modern['budget_to_genre_ratio'])
df_modern['competition_count'] = df_modern.groupby(['year', 'month', 'genre_clean'])['production_budget'].transform('count') - 1

# --- 5. DEFINE FEATURES ---
features = ['log_budget', 'month_sin', 'month_cos', 'genre_encoded', 'log_budget_to_genre_ratio', 'competition_count']

# --- 6. SPLIT DATA ---
X_train = df_modern.loc[train_mask, features]
y_train = df_modern.loc[train_mask, 'target']

X_test = df_modern.loc[test_mask, features]
y_test = df_modern.loc[test_mask, 'target']

print(f"Training samples: {X_train.shape[0]} | Testing samples: {X_test.shape[0]}")

# --- 7. TRAIN GRADIENT BOOSTING MODEL ---
# Gradient Boosting handles these complex interaction terms significantly better
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.05, max_depth=4, random_state=42)
model.fit(X_train, y_train)

# --- 8. EVALUATION ---
y_pred_log = model.predict(X_test)

y_test_original = np.expm1(y_test)
y_pred_original = np.expm1(y_pred_log)

print(f"\n--- Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test_original, y_pred_original):.2f}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test_original, y_pred_original)):.2f}")
print(f"R² Score (Log-scale): {r2_score(y_test, y_pred_log):.4f}")

print("\n--- Feature Importances ---")
importances = pd.Series(model.feature_importances_, index=features).sort_values(ascending=False)
print(importances)
```
