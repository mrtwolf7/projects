```python
import pandas as pd

df = pd.read_csv('raw_data/jfk-lax_2025.csv')
display(df.head(5))

```

```python
#reduce dimensionality - remove useless columns or columns with most NaNs

df_red = df[['Month','DayofMonth','DayOfWeek',
             'Reporting_Airline','Flight_Number_Reporting_Airline',
             'DepTime', 'DepDelay', 'ArrTime', 'ArrDelay'
            ]]
```

```python
df_red.head(30)
```

```python
# Dropping rows with NaNs
df_red_na = df_red.dropna()
len(df_red_na)
```

```python
df_red_na.to_csv('jfk-lax_2025_clean.csv')
```
